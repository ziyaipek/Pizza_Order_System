```python
import csv
import datetime

with open("Menu.txt", "w") as f:
    f.write("Pizzas :\n1: Classic\n2: Margherita\n3: Turkish Pizza\n4: Dominos Pizza\n5: Pepperoni\n6: Vegetarian\n\n")
    f.write("Toppings:\n11: Olive\n12: Mushroom\n13: Goat Cheese\n14: Meat\n15: Onion\n16: Corn\n\n")

# Creates the Pizza superclass

class Pizza:
  def __init__(self, description, cost):
        self.description = description
        self.cost = cost

  def get_description(self):
      return self.description

  def get_cost(self):
      return self.cost

# Creates Pizza subclasses

class Classic(Pizza):
  def __init__(self):
        super().__init__("Tomato Sauce, Mozzarella, Ham, Black Olives, Mushrooms", 10.99)
        

class Margherita(Pizza):

  def __init__(self):
    self.description = "San Marzano Tomatoes, Mozzarella Cheese, Fresh Basil, Salt, Extra-Virgin Olive Oil"
    self.cost = 30.0
  
class Turkish_Pizza(Pizza):

  def __init__(self):
    self.description = "Minced Meat, Onions, Garlic, Tomatoes, Red Peppers, Parsley, Chili Pepper, Paprika,"
    self.cost = 50.0

class Dominos_Pizza(Pizza):

  def __init__(self):
    self.description = "Tomato Sauce, Mozzarella, Bacon, Ham, Green Peppers, Onions, Mushrooms"
    self.cost = 45.0

class  Pepperoni(Pizza):

  def __init__(self):
    self.description = "Tomato Sauce, Mozzarella, Pepperoni"
    self.cost = 45.0

class Vegetarian(Pizza):

  def __init__(self):
    self.description = "Tomato Sauce, Mozzarella, Sweet Corn, Fresh Tomato, Green Peppers, Black Olives, Onions, Mushrooms"
    self.cost = 40.0

# Creates the Decorator superclass

class Decorator(Pizza):
  def __init__(self, toppings):
    self.toppings = toppings

  def get_cost(self):
       return self.toppings.get_cost() + Pizza.get_cost(self)

  def get_description(self):
       return self.toppings.get_description() + ' and extra topping ' + Pizza.get_description(self)

# Creates the Decorator subclass

class Olive(Decorator):

  def __init__(self, toppings):
    self.cost = 2.99
    self.description = "Olive"
    Decorator.__init__(self, toppings)

class Mushroom(Decorator):

  def __init__(self, toppings):
    self.cost = 3.99
    self.description = "Mushroom"
    Decorator.__init__(self, toppings)

class Goat_Cheese(Decorator):

  def __init__(self, toppings):
    self.cost = 3.99
    self.description = "Goat Cheese"
    Decorator.__init__(self, toppings)

class Meat(Decorator):

  def __init__(self, toppings):
    self.cost = 4.99
    self.description = "Meat"
    Decorator.__init__(self, toppings)

class Onion(Decorator):

  def __init__(self, toppings):
    self.cost = 2.99
    self.description = "Onion"
    Decorator.__init__(self, toppings)

class Corn(Decorator):

  def __init__(self, toppings):
    self.cost = 2.99
    self.description = "Corn"
    Decorator.__init__(self, toppings)

# Creates the main function for necessary operations

def main():
  with open("Menu.txt", "r") as customer_menu:
    for i in customer_menu:
      print(i)
  
  class_dict = {
                    1: Classic, 
                    2: Margherita, 
                    3: Turkish_Pizza, 
                    4: Dominos_Pizza, 
                    5: Pepperoni, 
                    6: Vegetarian,
                    11: Olive,
                    12: Mushroom,
                    13: Goat_Cheese,
                    14: Meat,
                    15: Onion,
                    16: Corn
                    }

  customer_choice = input("Please select your pizza: ")
  while customer_choice not in ["1", "2", "3", "4","5","6"]:
    customer_choice = input("You keyed wrong, please try again: ")
  order = class_dict[int(customer_choice)]()

  customer_choice = input("Please select your toppings: ")
  while customer_choice not in ["11", "12", "13", "14","15","16"]:
    customer_choice = input("You keyed wrong, please try again: ")
  order = class_dict[int(customer_choice)](order)

  info = "\nIngredients: " + order.get_description() + "\n= $" + str(order.get_cost()) + "\n"
  print(info)

 

# Creates an Order Information Card

  print("----------Order Information----------\n")
  name = input("Name: ")
  ID = input("ID Number: ")
  credit_card_number = input("Credit Card Number: ")
  credit_card_password = input("Credit Card Password: ")
  time_of_order = datetime.datetime.now()

  with open('Orders_Database.csv', 'a') as orders:
      orders = csv.writer(orders, delimiter=',')
      orders.writerow([name, ID, credit_card_number, credit_card_password, order.get_description() , order.get_cost(), time_of_order])
  print("\nThank you for your order!")

main()
```
